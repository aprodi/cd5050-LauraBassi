Ora importiamo i dati. Rispetto ai notebook precedenti proviamo a riunire i comandi necessari all'interno di una funzione python in modo da sperimentare un'altra modalità. 

Le funzioni in Python si definiscono con la parola chiave "def" seguita dal nome che vogliamo dare alla funzione e tra parentesi gli "argomenti" della funzione, cioè i parametri che forniamo alla funzione quando la chiamiamo, seguiti da ":" (dalla riga successiva cominciano le istruzioni che fanno parte della funzione). Come abbiamo visto nel caso del loop for, in Python il blocco delle istruzioni va "indentato", cioè allineato a una colonna successiva a quella di "def".

Per esempio, una funzione che prende in ingresso due numeri a e b e ne restituisce (return) la variabile sum a cui abbiamo assegnato la somma dei valori di a e di b, si scriverà:

     def my_function(a,b):
         sum = 0
         sum = a + b
         return sum


Un'altra cosa importante è che l'interprete di Python (cioè il programma che fa da "interprete" tra il nostro codice e il computer e gli dice cosa vogliamo che faccia) legge le istruzioni riga per riga, quindi dobbiamo sempre definire una funzione prima di poterla usare (subito prima oppure all'inizio del programma).

La funzione fn_carica_data() legge il csv direttamente dal repository di Github attraverso il suo indirizzo web e ci restituisce i dati in un Dataframe.
 
Il Dataframe è la struttura dati principale su cui Pandas internamente elabora i dati.

```python
def fn_carica_data(url):
    data = pd.read_csv(url)
    data = data.dropna(axis='columns', how ='all') #dropna è utilizzato per rimuovere i valori mancanti
    return data
```

Ora usiamo la funzione che abbiamo appena creato per caricare il dataset dei pinguini *penguins_lter.csv*. 

Qui potete caricare un qualsiasi dataset contenuto nella cartella di Github "Datasets", semplicemente sostituendo nell'indirizzo qui sotto il nome del database ".csv"

```python
df_pinguini = fn_carica_data("https://raw.githubusercontent.com/aprodi/cd5050-LiceoCastelnuovo/main/Datasets/penguins_lter.csv")
```

##Covarianza & Correlazione

“Covarianza” indica la direzione della relazione lineare tra le variabili. La "correlazione" d'altra parte misura sia la forza che la direzione della relazione lineare tra due variabili.

```python
# Covarianza
print('Covarianza:')
df_pinguini.cov()
```

```python
print('Correlazione:')
df_pinguini.corr()
```

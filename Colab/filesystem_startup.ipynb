Il _path_ e il nome del file (_filename_) sono spesso necessari all'interno di un programma o di un notebook _ipynb_ come argomenti di funzioni relative alla gestione dei file, per esempio per aprire un file e caricarlo in memoria, oppure per salvare i cambiamenti e chiudere il file. il _path_ e il _filename_ sono viste dal computer come stringhe di caratteri, e quindi possono essere contenute in una variabile di tipo stringa.


```python
# creiamo una variabile di nome path e assegnamole un percorso
path = '/content/drive/MyDrive/GitHub/cd5050-LauraBassi/Colab/'
print(path)
```

```python
# facciamo lo stesso per il nome del file ipynb
filename = 'Morgagni_M1.ipynb'
print(filename)
```

```python
# il riferimento assoluto di un file si compone del path e del filename
# possiamo "concatenare le stringe" (attaccare una stringa all'altra)
# relative al path e al filename con l'operatore '+'
absfile = path + filename
print(absfile)
```

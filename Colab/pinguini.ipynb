Ora importiamo i dati. Rispetto ai notebook precedenti non usiamo wget ma proviamo a farlo costruendo una funzione in modo da sperimentare un'altra modalità.

La funzione fn_carica_data() legge il csv e ci restituisce i dati in un Dataframe:

```python
def fn_carica_data(url):
    data = pd.read_csv(url)
    data = data.dropna(axis='columns', how ='all') #dropna è utilizzato per rimuovere i valori mancanti
    return data
```

Carichiamo il dataset dei pinguini *penguins_lter.csv*

```python
df_pinguini = fn_carica_data("https://raw.githubusercontent.com/aprodi/cd5050-LauraBassi/main/Datasets/penguins_lter.csv")
```

##Covarianza & Correlazione

“Covarianza” indica la direzione della relazione lineare tra le variabili. La "correlazione" d'altra parte misura sia la forza che la direzione della relazione lineare tra due variabili.

```python
# Covarianza
print('Covarianza:')
df_pinguini.cov()
```

```python
print('Correlazione:')
df_pinguini.corr()
```
